```python
import os
path = '/Users/nguye/Desktop/Sales-Reporting/'
# listdir : Hàm này chỉ liệt kê các tệp và thư mục có trong thư mục làm việc hiện tại.
list_files = os.listdir(path)
```

```python
import pandas as pd
frames = []
for file in list_files:
    filepath = path + file 
    frame = pd.read_csv(filepath)
    frames.append(frame)
# merge gop frame 12 thang thanh 1
df = pd.concat(frames)
```

```python
df
```

```python
#truy cap order date de lay gia tri thang 
df['Month'] = df['Order Date'].str[0:2] #df['Order Date'].str.slice(start=0,stop=2,step=1) 
df
```

```python
#lay cac unique value 
print(set(df['Month']))
```

```python
#loai bo cac gia tri nan va or , nan - cac gia tri bi mat
df = df.dropna(how = 'all')
```

```python
df[df['Month']== 'Or'] # vi khi gop da gop ca tieu de nen se xoa cac dong nay di Or in Order Date
```

```python
# cap nhap frame la cac dong khac 'Or'
df = df[df['Month']!= 'Or'] 
df
```

```python
#lay cac unique value 
print(set(df['Month']))
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.to_csv('sales2019.csv', index = False)
```

## Tim thang co doanh so tot nhat , doanh so thang la bao nhieu What was the best month for sales? How much was earned that month?

```python
# convert object dtypes to numeric
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], downcast = 'integer')
df['Price Each'] = pd.to_numeric(df['Price Each'], downcast = 'integer')
```

```python
df['Sales']= df['Quantity Ordered']*df['Price Each']
df
```

```python
# chuyen cot sales den vi tri sau cot price each
moving_column = df.pop('Sales')
df.insert(4,'Sales',moving_column)
df
```

```python
sales_values = df.groupby(['Month']).sum()['Sales']
sales_values.max()
```

```python
import matplotlib.pyplot as plt
months = range(1,13)
plt.bar(months, sales_values)
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

## Thanh pho nao co danh so cao nhat What city has the best sales

```python
df
```

```python
# lay ten thanh pho tu cot Purchase Address 
'''def getCity(city):
    return city.split(',')[1]
getCity('840 Highland St, Los Angeles, CA 90001')'''
address_to_city = lambda city:city.split(',')[1]
```

```python
df['City'] = df['Purchase Address'].apply(address_to_city)
df
```

```python
sales_city = df.groupby(['City']).sum()['Sales']
cities = list(sales_city.index)
cities
```

```python

plt.bar(cities, sales_city)
plt.xticks(cities, rotation = -45, size = 8)
plt.xlabel( 'Cities')
plt.ylabel('Sales')
plt.show()
```
